```python
from google.colab import files
import pandas as pd

# Step 1: Upload the Dataset
print("Please upload your dataset:")
uploaded = files.upload()

# Step 2: Load the Dataset
# Get the filename from the uploaded dictionary
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)

# Step 3: Display the First Few Rows
print("\nFirst few rows of the dataset:")
print(data.head())

# Step 4: General Information
print("\nDataset Structure:")
print(data.info())

print("\nNumber of Missing or Zero Values:")
# Count zeros in numerical columns except for Outcome (Outcome=0 is valid)
zero_counts = (data == 0).sum()
print(zero_counts)

# Identify columns with missing or zero values that might be problematic
columns_to_check = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
print("\nPotentially problematic zero counts:")
print(zero_counts[columns_to_check])

# Step 5: Descriptive Analysis
print("\nDescriptive Statistics:")
print(data.describe())

# Look for irregularities
print("\nIrregularities Check:")
for column in columns_to_check:
    print(f"{column}: Min = {data[column].min()}, Max = {data[column].max()}")
```
