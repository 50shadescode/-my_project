```python
from google.colab import files

uploaded = files.upload()
import pandas as pd

# Read the uploaded file
df = pd.read_csv('Tunisair_flights_dataset(1).csv')  # Adjust the filename if needed

# Display the first few rows
df.head()
```

```python
# Overview of the dataset
print("Dataset Info:")
print(df.info())

# Display the first 5 rows of the dataset
print("\nFirst 5 rows of the dataset:")
print(df.head())

# Display the last 5 rows of the dataset
print("\nLast 5 rows of the dataset:")
print(df.tail())
# Check for missing values
print("\nMissing values in the dataset:")
print(df.isnull().sum())
# Summary statistics for numerical columns
print("\nSummary statistics for numerical columns:")
print(df.describe())
# List unique values for each categorical column
categorical_columns = df.select_dtypes(include=['object']).columns

print("\nUnique values in categorical columns:")
for column in categorical_columns:
    print(f"{column}: {df[column].unique()}")
```

```python
import matplotlib.pyplot as plt  # Add this import statement

# Plot a histogram for the 'Arrival delay' column
df['Arrival delay'].hist()
plt.title('Distribution of Arrival Delays')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Frequency')
plt.show()
```
